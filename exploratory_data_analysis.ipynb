#### Step 7: Analyzing Programming Language Trends by Region


Analyze the popularity of programming languages by region. Use the `LanguageHaveWorkedWith` column to investigate which languages are most used in different regions.

- Filter data by country or region.

- Visualize the top programming languages by region with a bar plot or heatmap.


```python
df_lang = df[['Country', 'LanguageHaveWorkedWith']].dropna()

df_lang = df_lang.assign(Language=df_lang['LanguageHaveWorkedWith'].str.split(';')).explode('Language')
```

```python
top_countries = df_lang['Country'].value_counts().head(5).index
df_lang_top = df_lang[df_lang['Country'].isin(top_countries)]

language_counts = df_lang_top.groupby(['Country', 'Language']).size().reset_index(name='Count')
top_languages = language_counts.sort_values(['Country', 'Count'], ascending=[True, False]).groupby('Country').head(5)
```

```python

plt.figure(figsize=(12, 6))
sns.barplot(
    data=top_languages,
    x='Language',
    y='Count',
    hue='Country'
)

plt.title('Top Programming Languages by Country')
plt.tight_layout()
plt.show()
```

#### Step 9: Educational Background and Employment Type


Explore how educational background (`EdLevel`) relates to employment type (`Employment`). Use cross-tabulation and visualizations to understand if higher education correlates with specific employment types.


```python
ed_employment_pct = pd.crosstab(
    df['EdLevel'],
    df['Employment'],
    normalize='index'
)

ed_employment_pct.plot(
    kind='bar',
    stacked=True,
    figsize=(16, 6), 
    colormap='tab20'
)
plt.xticks(rotation=90)
plt.title('Education Level vs Employment Type')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Employment')
plt.show()
```
