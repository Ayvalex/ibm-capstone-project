### 1. Load the Dataset

### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
list(df['Country'].unique())
```

```python
df['Country'].value_counts().tail(30)
```

Inconsistencies found: mix of long and short country naming conventions, multiple naming conventions for the Koreas ("South Korea", "Republic of Korea", "Democratic People's Republic of Korea"), a space in "Viet Nam", and a irrelevant entry "Nomadic". 

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
country_mapping = {
    'Republic of Korea': 'South Korea',
    "Democratic People's Republic of Korea": 'North Korea',
    'Viet Nam': 'Vietnam',
    'United States of America': 'USA',
    'United Kingdom of Great Britain and Northern Ireland': 'UK',
    'Russian Federation': 'Russia',
    'Iran, Islamic Republic of...': 'Iran',
    'Venezuela, Bolivarian Republic of...': 'Venezuela'
}

df['Country'] = df['Country'].replace(country_mapping)
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
df['Employment'] = df['Employment'].str.strip()

employment_dummies = pd.get_dummies(df['Employment'])

df_clean = pd.concat([df, employment_dummies], axis=1)

df_clean = df_clean.drop(columns=['Employment'])

df_clean
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
df.isna().sum().sort_values(ascending=False)
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
num_cols = df_clean.select_dtypes(include=['int64', 'float64']).columns
df_clean[num_cols] = df_clean[num_cols].fillna(df_clean[num_cols].median())
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
cat_cols = df_clean.select_dtypes(include='object').columns

for col in cat_cols:
    mode_value = df_clean[col].mode()[0]
    df_clean[col] = df_clean[col].fillna(mode_value)
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
df_clean = df_clean.dropna(subset=['ConvertedCompYearly'])
min_value = df_clean['ConvertedCompYearly'].min()
max_value = df_clean['ConvertedCompYearly'].max()

df_clean['ConvertedCompYearly_MinMax'] = (df_clean['ConvertedCompYearly'] - min_value) / (max_value - min_value)
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
import numpy as np

print((df_clean['ConvertedCompYearly'] <= 0).sum())  
df_clean['ConvertedCompYearly_Log'] = np.log(df_clean['ConvertedCompYearly'])
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
non_numeric = df_clean[~df_clean['YearsCodePro'].astype(str).str.isdigit()]

non_numeric['YearsCodePro'].unique()
```

```python
df_clean['YearsCodePro'] = df_clean['YearsCodePro'].replace({
    'Less than 1 year': 0.5,
    'More than 50 years': 50
})
df_clean['YearsCodePro'] = pd.to_numeric(df_clean['YearsCodePro'])
```

```python
df_clean['ExperienceLevel'] = pd.cut(
    df_clean['YearsCodePro'],
    bins=[-1, 2, 5, 10, np.inf],
    labels=['Beginner', 'Intermediate', 'Experienced', 'Advanced/Expert']
)
```
